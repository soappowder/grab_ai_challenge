## data dictionary

- bookingID: trip id
- Accuracy: accuracy inferred by GPS in meters
- Bearing: GPS bearing
- acceleration: accelerometer reading (m/s2)
- gyro: gyroscope reading (rad/s)
- second: time of the record by number of seconds
- Speed: speed measured by GPS in m/s

## load data

```python
def load_data(kind):
    path ='data/safety/'+ kind
    all_files = glob.glob(os.path.join(path, "*.csv")) 
    features = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
    return features
```

```python
features = load_data(kind="features")
```

```python
labels = load_data(kind="labels")
```

## scaling

- the features have many outliers
- these outliers are important (indicative of dangerous driving behaviour)
- hence, we do not want to clip the data

- StandardScaler and MinMaxScaler are sensitive to the presence of outliers
- hence, we'll use RobustScaler

```python
features_df_clean_scaled = features_df.copy()
```

```python
# shifting acceleration_y by gravitational force g = 9.81

# shift
g = 9.81
features_df_clean_scaled['acceleration_y_shifted'] = features_df_clean_scaled['acceleration_y'] - g
# drop
features_df_clean_scaled.drop('acceleration_y', axis=1, inplace=True)
# rename
features_df_clean_scaled.rename(columns={'acceleration_y_shifted':'acceleration_y'}, inplace=True)
```

```python
# Robust Scaling for GPS data
scaled_cols = ['Accuracy', 'Bearing','Speed']
scaler = RobustScaler(copy=True, quantile_range=(25.0, 75.0), with_centering=True, with_scaling=True)
features_df_clean_scaled[scaled_cols] = scaler.fit_transform(features_df_clean_scaled[scaled_cols])
```

```python
# Robust Scaling for acceleration
# since acceleration_x looks like in the right scale, we don't want to scale acceleration again
# since 0 acceleration has special meaning, we don't want to center acceleration
```

```python
# Robust Scaling for gyro
# since 0 gyro has special meaning, we don't want to center gyro (it's already quite centered)
# with_centering = False
scaled_cols = ['gyro_x', 'gyro_y','gyro_z']
scaler = RobustScaler(copy=True, quantile_range=(25.0, 75.0), with_centering=False, with_scaling=True)
features_df_clean_scaled[scaled_cols] = scaler.fit_transform(features_df_clean_scaled[scaled_cols])
```

```python
# box plot after scaling
plt.figure()
ax1 = plt.subplot(331)
features_df_clean_scaled.Accuracy.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax1)
ax2 = plt.subplot(332)
features_df_clean_scaled.Bearing.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax2)
ax3 = plt.subplot(333)
features_df_clean_scaled.Speed.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax3)
ax4 = plt.subplot(334)
features_df_clean_scaled.acceleration_x.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax4)
ax5 = plt.subplot(335)
features_df_clean_scaled.acceleration_y.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax5)
ax6 = plt.subplot(336)
features_df_clean_scaled.acceleration_z.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax6)
ax7 = plt.subplot(337)
features_df_clean_scaled.gyro_x.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax7)
ax8 = plt.subplot(338)
features_df_clean_scaled.gyro_y.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax8)
ax9 = plt.subplot(339)
features_df_clean_scaled.gyro_z.plot(kind='box', figsize=[10,10], showfliers=False, ax=ax9)
```
